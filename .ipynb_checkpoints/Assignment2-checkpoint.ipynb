```python
# imports, setup
import pandas as pd
import numpy as np

from sklearn.model_selection import KFold, cross_validate, GridSearchCV
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, r2_score, make_scorer

import matplotlib.pyplot as plt

# load data
df = pd.read_csv("steel.csv")

# features/target
X = df.drop("tensile_strength", axis=1)
y = df["tensile_strength"]

scoring = {
    "R2": make_scorer(r2_score),
    "MAE": make_scorer(mean_absolute_error)
}

kfold = KFold(n_splits=10, shuffle=True, random_state=1)
```

```python
#SVR
```

```python
svr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

svr_default_scores = cross_validate(
    svr_pipeline, X, y, cv=kfold,
    scoring=scoring, return_train_score=True
)

print("SVR Default Results")
print("Average Train R²:", svr_default_scores['train_R2'].mean())
print("Average Test R²:", svr_default_scores['test_R2'].mean())
print("Average Train MAE:", svr_default_scores['train_MAE'].mean())
print("Average Test MAE:", svr_default_scores['test_MAE'].mean())

svr_params = {
    "svr__C": [0.1, 1, 10, 100],
    "svr__gamma": ["scale", 0.01, 0.1, 1]
}

svr_grid = GridSearchCV(
    svr_pipeline,
    param_grid=svr_params,
    scoring="neg_mean_absolute_error",
    cv=kfold,
    n_jobs=-1
)

svr_grid.fit(X, y)

print("\nBest SVR hyperparameters:", svr_grid.best_params_)
svr_best_scores = cross_validate(
    svr_grid.best_estimator_, X, y, cv=kfold,
    scoring=scoring, return_train_score=True
)

print("\nSVR Tuned Results")
print("Average Train R²:", svr_best_scores['train_R2'].mean())
print("Average Test R²:", svr_best_scores['test_R2'].mean())
print("Average Train MAE:", svr_best_scores['train_MAE'].mean())
print("Average Test MAE:", svr_best_scores['test_MAE'].mean())


# tuning plots

svr_results = pd.DataFrame(svr_grid.cv_results_)
svr_results['MAE'] = -svr_results['mean_test_score']  # convert to positive MAE

# mae heatmap
pivot_svr = svr_results.pivot(index='param_svr__C', columns='param_svr__gamma', values='MAE')

plt.figure(figsize=(8, 5))
plt.title("SVR Tuning: Mean MAE (10-fold CV)")
im = plt.imshow(pivot_svr, cmap="viridis", aspect='auto')
plt.colorbar(im, label="Mean Absolute Error")
plt.xlabel("gamma")
plt.ylabel("C")
plt.xticks(range(len(pivot_svr.columns)), pivot_svr.columns)
plt.yticks(range(len(pivot_svr.index)), pivot_svr.index)
plt.show()

# line plot of c for each gamma
plt.figure(figsize=(8, 5))
for gamma in pivot_svr.columns:
    plt.plot(pivot_svr.index, pivot_svr[gamma], marker="o", label=f"gamma={gamma}")

plt.title("SVR Tuning: MAE vs C for Different Gamma Values")
plt.xlabel("C")
plt.ylabel("Mean CV MAE")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#KNNR
```

```python
knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])

knn_default_scores = cross_validate(
    knn_pipeline, X, y, cv=kfold,
    scoring=scoring, return_train_score=True
)

print("\nKNN Default Results")
print("Average Train R²:", knn_default_scores['train_R2'].mean())
print("Average Test R²:", knn_default_scores['test_R2'].mean())
print("Average Train MAE:", knn_default_scores['train_MAE'].mean())
print("Average Test MAE:", knn_default_scores['test_MAE'].mean())

knn_params = {
    "knn__n_neighbors": [2, 3, 5, 7, 10, 12, 15],
    "knn__weights": ["uniform", "distance"]
}

knn_grid = GridSearchCV(
    knn_pipeline,
    param_grid=knn_params,
    scoring="neg_mean_absolute_error",
    cv=kfold,
    n_jobs=-1
)

knn_grid.fit(X, y)

print("\nBest KNN hyperparameters:", knn_grid.best_params_)

knn_best_scores = cross_validate(
    knn_grid.best_estimator_, X, y, cv=kfold,
    scoring=scoring, return_train_score=True
)

print("\nKNN Tuned Results")
print("Average Train R²:", knn_best_scores['train_R2'].mean())
print("Average Test R²:", knn_best_scores['test_R2'].mean())
print("Average Train MAE:", knn_best_scores['train_MAE'].mean())
print("Average Test MAE:", knn_best_scores['test_MAE'].mean())


# tuning plots:

knn_results = pd.DataFrame(knn_grid.cv_results_)
knn_results['MAE'] = -knn_results['mean_test_score']  # convert to positive MAE

# line plot, weights
plt.figure(figsize=(8, 5))

for w in knn_results['param_knn__weights'].unique():
    subset = knn_results[knn_results['param_knn__weights'] == w]
    plt.plot(subset['param_knn__n_neighbors'], subset['MAE'], marker="o", label=f"Weights={w}")

plt.title("KNN Tuning: MAE vs Number of Neighbors")
plt.xlabel("n_neighbors")
plt.ylabel("Mean CV MAE")
plt.legend()
plt.grid(True)
plt.show()

# bar chart comparing weights for diff neighbors
pivot_knn = knn_results.pivot(index='param_knn__n_neighbors', columns='param_knn__weights', values='MAE')

pivot_knn.plot(kind="bar", figsize=(8, 5))
plt.title("KNN Tuning: Mean MAE (10-fold CV)")
plt.ylabel("Mean Absolute Error")
plt.xlabel("n_neighbors")
plt.legend(title="weights")
plt.xticks(rotation=0)
plt.grid(False)
plt.show()

```

```python
#results
```

```python
# fit best models for comparison
# best svr model
best_svr = svr_grid.best_estimator_
best_svr.fit(X, y)

# best knn model
best_knn = knn_grid.best_estimator_
best_knn.fit(X, y)

# generate cross-val scores again ONLY FOR TEST SETS
svr_best_scores = cross_validate(best_svr, X, y, cv=kfold, scoring=scoring)
knn_best_scores = cross_validate(best_knn, X, y, cv=kfold, scoring=scoring)

# compute averages
model_compare = pd.DataFrame({
    "Model": ["SVR (Default)", "KNN (Default)", "SVR (Tuned)", "KNN (Tuned)"],
    "Test_R2": [
        svr_default_scores['test_R2'].mean(),
        knn_default_scores['test_R2'].mean(),
        svr_best_scores['test_R2'].mean(),
        knn_best_scores['test_R2'].mean()
    ],
    "Test_MAE": [
        svr_default_scores['test_MAE'].mean(),
        knn_default_scores['test_MAE'].mean(),
        svr_best_scores['test_MAE'].mean(),
        knn_best_scores['test_MAE'].mean()
    ]
})

model_compare

# mae plot

plt.figure(figsize=(8, 5))
plt.bar(model_compare["Model"], model_compare["Test_MAE"])
plt.title("Model Comparison: Mean Test MAE (10-fold CV)")
plt.ylabel("MAE (Lower is Better)")
plt.xticks(rotation=10)
plt.grid(axis='y')
plt.show()

# r^2

plt.figure(figsize=(8, 5))
plt.bar(model_compare["Model"], model_compare["Test_R2"], color="lightgreen")
plt.title("Model Comparison: Mean Test R² (10-fold CV)")
plt.ylabel("R² Score (Higher is Better)")
plt.xticks(rotation=10)
plt.grid(axis='y')
plt.show()

# r^2, mae together

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# mae
ax[0].bar(model_compare["Model"], model_compare["Test_MAE"], color="skyblue")
ax[0].set_title("Comparison of Mean Test MAE")
ax[0].set_ylabel("MAE (Lower Better)")
ax[0].tick_params(axis='x', rotation=10)
ax[0].grid(axis='y')

# r^2
ax[1].bar(model_compare["Model"], model_compare["Test_R2"], color="lightgreen")
ax[1].set_title("Comparison of Mean Test R²")
ax[1].set_ylabel("R² (Higher Better)")
ax[1].tick_params(axis='x', rotation=10)
ax[1].grid(axis='y')

plt.tight_layout()
plt.show()

```

```python
#svr default vs tuned plot
svr_plot_data = pd.DataFrame({
    "Model": ["SVR Default", "SVR Tuned"],
    "Test_R2": [
        svr_default_scores['test_R2'].mean(),
        svr_best_scores['test_R2'].mean()
    ],
    "Test_MAE": [
        svr_default_scores['test_MAE'].mean(),
        svr_best_scores['test_MAE'].mean()
    ]
})

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# mae
ax[0].bar(svr_plot_data["Model"], svr_plot_data["Test_MAE"], color=["steelblue", "orange"])
ax[0].set_title("SVR: Test MAE (Lower Better)")
ax[0].set_ylabel("Mean Absolute Error")
ax[0].grid(axis='y')

# r^2
ax[1].bar(svr_plot_data["Model"], svr_plot_data["Test_R2"], color=["lightgreen", "salmon"])
ax[1].set_title("SVR: Test R² (Higher Better)")
ax[1].set_ylabel("R² Score")
ax[1].grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict

#cross_val_predict to get predictions for the full dataset
knn_best_model = knn_grid.best_estimator_
y_pred_knn_best = cross_val_predict(knn_best_model, X, y, cv=kfold)

# calc r^2 for tuned model
r2_knn_best = r2_score(y, y_pred_knn_best)
mae_knn_best = mean_absolute_error(y, y_pred_knn_best)

knn_plot_data = pd.DataFrame({
    "Model": ["KNN Default", "KNN Tuned"],
    "Test_R2": [
        knn_default_scores['test_R2'].mean(),
        r2_knn_best
    ],
    "Test_MAE": [
        knn_default_scores['test_MAE'].mean(),
        mae_knn_best
    ]
})

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# mae
ax[0].bar(knn_plot_data["Model"], knn_plot_data["Test_MAE"], color=["steelblue", "orange"])
ax[0].set_title("KNN: Test MAE (Lower Better)")
ax[0].set_ylabel("Mean Absolute Error")
ax[0].grid(axis='y')

# r^2
ax[1].bar(knn_plot_data["Model"], knn_plot_data["Test_R2"], color=["lightgreen", "salmon"])
ax[1].set_title("KNN: Test R² (Higher Better)")
ax[1].set_ylabel("R² Score")
ax[1].grid(axis='y')

plt.tight_layout()
plt.show()
```
